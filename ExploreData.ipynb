```python
import pandas as pd
import numpy as np
import os
```

```python
raw_data_path = os.path.join(os.path.pardir,'data','raw')
train_data_path = os.path.join(raw_data_path,'train.csv')
test_data_path = os.path.join(raw_data_path,'test.csv')
print(raw_data_path,train_data_path,test_data_path)
```

```python
train_df = pd.read_csv(train_data_path,index_col='PassengerId')
test_df = pd.read_csv(test_data_path,index_col='PassengerId')
```

```python
type(train_df)
```

```python
#Adding new column Survived with default value i.e. -888
test_df['Survived'] = -888
```

```python
df = pd.concat((train_df,test_df),sort=False)
```

```python
gender = 'female'
male_passenger = df.loc[df.Sex==gender]
print('number of {g} passengees:{a}'.format(a=len(male_passenger),g=gender))
```

```python
#filering with & and | condition
gender = 'female'
klass = 1
male_passenger = df.loc[(df.Sex==gender) | (df.Pclass==klass)]
print('number of {g} passengers in {k} class is :{a}'.
      format(a=len(male_passenger),g=gender,k=klass))
```
