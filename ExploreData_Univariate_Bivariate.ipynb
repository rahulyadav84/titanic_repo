# Grouping and Aggregations

```python
df.groupby(by=["Pclass"]).agg({'Fare':'mean','Age':'median'})
```

```python
df.Fare.max() - df.Fare.min()
aggregations = {'Age':{'mean_val':'mean','median_val':'median','age_range': lambda x:max(x)-min(x)}}
df.groupby(by=["Pclass"]).agg(aggregations)
    
```

```python
df.groupby('Pclass').Fare.median()
```
